#  Gráficos de Barras e Linhas
***
Os gráficos que combinam barras e linhas em um mesmo gráfico são ferramentas valiosas na visualização de dados complexos e na análise de múltiplas variáveis simultaneamente. Essa combinação de elementos gráficos permite que informações complementares sejam apresentadas de forma clara e concisa, facilitando a compreensão e interpretação dos dados.

A importância desses gráficos está em sua capacidade de fornecer uma visão abrangente e comparativa das informações. As barras são especialmente úteis para representar dados categóricos ou agrupados, permitindo visualizar e comparar diferentes categorias ou grupos de maneira eficiente. Por outro lado, as linhas são ideais para representar tendências ou relações ao longo do tempo, mostrando a evolução e a interconexão entre diferentes variáveis.

A combinação de barras e linhas em um único gráfico nos permite analisar simultaneamente a **distribuição de categorias e a evolução temporal, destacando padrões, relações e discrepâncias**. Isso é particularmente útil em situações onde desejamos comparar categorias diferentes ao longo do tempo ou analisar a interação entre duas variáveis relacionadas.

Existem várias situações no mundo real em que a combinação de gráficos de barras e linhas em um mesmo gráfico pode ser útil.

<input type="checkbox" checked disabled> **Análise de vendas por região ao longo do tempo:** O gráfico pode mostrar as vendas totais de cada região em barras, enquanto a linha representa a média geral de vendas. Isso permite comparar o desempenho de cada região com a média e identificar tendências ou discrepâncias.

<input type="checkbox" checked disabled > **Monitoramento do crescimento de uma empresa:** As barras podem representar o crescimento percentual de receita ou lucro por trimestre, enquanto a linha mostra a tendência geral de crescimento ao longo do tempo. Isso permite avaliar o desempenho trimestral em relação à trajetória de crescimento.

<input type="checkbox" checked disabled> **Acompanhamento de indicadores sociais:** Por exemplo, o gráfico pode apresentar as taxas de desemprego por faixa etária em barras, enquanto a linha mostra a taxa geral de desemprego. Isso ajuda a identificar grupos populacionais que estão enfrentando maiores desafios de emprego em comparação com a média.

<input type="checkbox" checked disabled> **Análise de gastos de marketing por canal:** As barras podem representar os gastos de marketing em cada canal (TV, rádio, online, etc.), enquanto a linha mostra o retorno sobre o investimento (ROI) ao longo do tempo. Isso permite identificar os canais mais eficazes em termos de ROI e comparar os gastos com os resultados obtidos.

<input type="checkbox" checked disabled> **Acompanhamento de métricas de desempenho de projetos:** O gráfico pode mostrar o progresso de diferentes métricas-chave em barras (por exemplo, custo, tempo, qualidade) e a linha pode representar a meta estabelecida. Isso permite visualizar o desempenho real em relação às metas estabelecidas.

### Importação das bibliotecas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Dados de exemplo

```python
# Dados fictícios
anos = [2018, 2019, 2020, 2021, 2022]
receita_total = [50000, 60000, 70000, 90000, 95000]
receita_produto1 = [20000, 22000, 25000, 28000, 30000]
receita_produto2 = [30000, 38000, 45000, 62000, 65000]
```

```python

# Criando DataFrame
data = {
    'Ano': anos,
    'Receita Produto 1': receita_produto1,
    'Receita Produto 2': receita_produto2,
    'Receita Total': receita_total
}
df = pd.DataFrame(data)


# Criando a figura e os eixos
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plotando as barras
df[['Ano','Receita Produto 1', 'Receita Produto 2']].plot(x='Ano', 
                                                          kind='bar', 
                                                          stacked=True, 
                                                          ax=ax1, 
                                                          color=['blue', 'green'], 
                                                          alpha=0.5)

# Configurando os eixos das barras
ax1.set_xlabel('Ano')
ax1.set_ylabel('Receita por Produto')

# Criando o eixo secundário
ax2 = ax1.twinx()

# Plotando o gráfico de linha
df['Receita Total'].plot(kind='line', marker='o', color='red', ax=ax1, legend=True)
plt.plot(anos, receita_total, color='red', marker='o', label='Receita Total')

# Configurando o eixo secundário
ax2.set_ylabel('Receita Total')

# Configurando o título
plt.title('Receita Total e por Produto ao longo dos Anos')

# Ajustando o layout
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
