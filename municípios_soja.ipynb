MUNICÍPIOS COM PRODUÇÃO DE SOJA

```python
import urllib3
import json

def get_data(url):
  http = urllib3.PoolManager()
  r = http.request('GET', url)
  data = r.data
  res = json.loads(data)
  return res

populacao = get_data('https://servicodados.ibge.gov.br/api/v3/agregados/6579/periodos/2021/variaveis/9324?localidades=N6[all]')
area_soja = get_data('https://servicodados.ibge.gov.br/api/v3/agregados/1612/periodos/2021/variaveis/109?localidades=N6[all]&classificacao=81[2713]')
```

AGORA FAREI UMA LISTA ORDENADA DA POPULAÇÃO COM O MUNICÍPIO

```python
for j in range(len(pop_pop)):
  ind_menor = j
  val_menor = pop_pop[j]
  for i in range(j, len(pop_pop)):
    if pop_pop[i] < val_menor:
      ind_menor = i
      val_menor - pop_pop[i]

  atual_j = pop_pop[j]
  pop_pop[j] = val_menor
  pop_pop[ind_menor] = atual_j

  atual_j = pop_mun[j]
  pop_mun[j] = val_menor
  pop_mun[ind_menor] = atual_j

  atual_mun = pop_mun[j]
  menor_mun = pop_mun[ind_menor]
  pop_mun[j] = menor_mun
  pop_mun[ind_menor] = atual_mun
```
